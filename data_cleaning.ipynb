```python
def transform(column):
    
    qnum = column.split(".")[0]
    subscale = column.split(":")[0].split(".")[-1].strip().title().replace(" ", "")
    if construct.lower() != "subjectivequality":
        title = f"Q{qnum}{construct}{subscale}"
    else:
        title = f"Q{qnum}{construct}"
    return title
    

f = data.copy()

# section A: Engagement Q1-Q5
construct="Engagement"
f.rename(columns=dict(zip(f.columns[2:7], f.columns[2:7].map(transform))), inplace=True)

# section B: Engagement Q6-Q9
construct="Functionality"
f.rename(columns=dict(zip(f.columns[7:11], f.columns[7:11].map(transform))), inplace=True)

# section C: Aesthetics Q10-12
construct="Aesthetics"
f.rename(columns=dict(zip(f.columns[11:14], f.columns[11:14].map(transform))), inplace=True)

# section D: Information Q13-16
construct="Information"
f.rename(columns=dict(zip(f.columns[14:18], f.columns[14:18].map(transform))), inplace=True)

# section E: Subjective Quality Q17-20
construct="SubjectiveQuality"
f.rename(columns=dict(zip(f.columns[18:22], f.columns[18:22].map(transform))), inplace=True)
```

```python
data.loc[:, "BoolDiet"] = data.loc[:, "BoolDiet"].apply(lambda x: 1 if x == "Yes" else 0)
```

```python
data.loc[:, "BoolFoodAllergens"] = data.loc[:, "BoolFoodAllergens"].apply(lambda x: 1 if x == "Yes" else 0)
```

```python
data.loc[:, "Sex"] = data.loc[:, "Sex"].apply(lambda x: 1 if x == "Male" else 0)
```

```python
data.loc[:, "CivilStatus"] = data.loc[:, "CivilStatus"].replace("Live in", "Married").replace("Widowed ", "Single")
data.loc[:, "CivilStatus"] = data.loc[:, "CivilStatus"].replace("Single", 0).replace("Married", 1)
```

```python
data.loc[:, "YearsOfCooking"] = data.loc[:, "YearsOfCooking"].replace("Less than a year", 1
                          ).replace("1 to 5 years", 2
                          ).replace("6 to 10 years", 3
                          ).replace("11 and above", 4)
```

```python
data.loc[:, "MonthlyIncome"] = data.loc[:, "MonthlyIncome"].replace("low", 1).replace("medium", 2).replace("high", 3)
data.head()
```
