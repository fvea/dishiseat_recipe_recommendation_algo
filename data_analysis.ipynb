```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel("dishiseat_cleaned.xlsx")
data.head()
```

```python
engagement = [
    "Q1EngagementEntertainment", 
    "Q2EngagementInterest", 
    "Q3EngagementCustomization",
    "Q4EngagementInteractivity",
    "Q5EngagementTargetGroup"
]

engagement_data = data.loc[:, engagement].copy()
```

```python
engagement_data.describe()
```

```python
print(engagement_data.mean(axis=1).describe())
print()
print(engagement_data.mean(axis=1).skew())
```

```python
functionality = [
    "Q6FunctionalityPerformance",
    "Q7FunctionalityEaseOfUse",
    "Q8FunctionalityNavigation",
    "Q9FunctionalityGesturalDesign"
]

funct_data = data.loc[:, functionality].copy()
funct_data.describe()
```

```python
print(funct_data.mean(axis=1).describe())
print()
print(funct_data.mean(axis=1).skew())
```

```python
aesthetics = [
    "Q10AestheticsLayout",
    "Q11AestheticsGraphics",
    "Q12AestheticsVisualAppeal",
]

aes_data = data.loc[:, aesthetics].copy()
aes_data.describe()
```

```python
print(aes_data.mean(axis=1).describe())
print()
print(aes_data.mean(axis=1).skew())
```

```python
information = [
    "Q13InformationQualityOfInformation",
    "Q14InformationQuantityOfInformation",
    "Q15InformationVisualInformation",
    "Q16InformationCredibilityOfSource"
]

info_data = data.loc[:, information].copy()
info_data.describe()
```

```python
print(info_data.mean(axis=1).describe())
print()
print(info_data.mean(axis=1).skew())
```

```python
subj_quality = [
    "Q17SubjectiveQuality",
    "Q18SubjectiveQuality",
    "Q19SubjectiveQuality",
    "Q20SubjectiveQuality" 
]

subj_data = data.loc[:, subj_quality].copy()
subj_data.describe()
```

```python
print(subj_data.mean(axis=1).describe())
print()
print(subj_data.mean(axis=1).skew())
```

```python
measures = pd.DataFrame()
measures["Engagement"] = engagement_data.mean(axis=1)
measures["Functionality"] = funct_data.mean(axis=1)
measures["Aesthetics"] = aes_data.mean(axis=1)
measures["Information"] = info_data.mean(axis=1)
measures["SubjectiveQuality"] = subj_data.mean(axis=1)
measures.head()
```

```python
demog_data = data.loc[:, ["Sex", "Age", "CivilStatus", "BoolDiet", 
             "BoolFoodAllergens", "YearsOfCooking", 
             "CookingSkills", "MonthlyIncome"]].copy()

demog_data.head()
```

```python
df = pd.concat([measures, demog_data], axis=1)
df.head()
```

```python
df.loc[:, "CivilStatus"] = df.loc[:, "CivilStatus"].replace("Live in", "Married").replace("Widowed ", "Single")
df.loc[:, "CivilStatus"] = df.loc[:, "CivilStatus"].replace("Single", 0).replace("Married", 1)
```

```python
df.loc[:, "YearsOfCooking"] = df.loc[:, "YearsOfCooking"].replace("Less than a year", 1
                          ).replace("1 to 5 years", 2
                          ).replace("6 to 10 years", 3
                          ).replace("11 and above", 4)
df
```

```python
df.loc[:, "MonthlyIncome"] = df.loc[:, "MonthlyIncome"].replace("low", 1).replace("medium", 2).replace("high", 3)
df
```

```python
df.iloc[:, 5:].corrwith(
    df.loc[:, ["Engagement", "Functionality", "Aesthetics", "Information", "SubjectiveQuality"]].mean(axis=1), 
    method="pearson")
```

```python
df.loc[:, ["Engagement", "Functionality", "Aesthetics", "Information"]].mean(axis=1)
```

```python
overall_quality = df.iloc[:, :5].mean(axis=1)
df.iloc[:, 5:].corrwith(overall_quality)
```
